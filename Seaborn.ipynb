# 1. Relational Plots


*   Scatter Plot
*   Line Plot



```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tip = sns.load_dataset('tips')
```

```python
tip
```

Scatter Plot

```python
# axes level function
sns.scatterplot(data=tip,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
# figure level function
sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter',hue='sex')
```

Line Plot

```python
gap = px.data.gapminder()
gap
```

```python
india = gap[gap['country']=='India']
```

```python
# axes level function
sns.lineplot(data=india,x='year',y='lifeExp')
```

```python
# figure level function
sns.relplot(data=india,x='year',y='lifeExp',kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(data=temp_df,kind='line',x='year',y='lifeExp',hue='country')
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country',style='continent')
```

Facet Plot

```python
# facet plot -> it will work only on figure level function
sns.relplot(data=tip,kind='scatter',x='total_bill',y='tip',col='day',row='sex')
```

```python
# colwrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

# 2. Distribution Plots

*   histplot
*   kdeplot

*   rugplot






```python
import seaborn as sns
```

Histplot

```python
#figure level function ->displot
#axes level function ->histplot ->kdeplot ->rugplot
sns.histplot(data=tip,x='total_bill')
```

```python
sns.displot(data=tip,x='total_bill',bins=5,kind='hist')
```

```python
sns.displot(data=tip,x='tip', kind='hist',hue='sex')
```

```python
sns.histplot(data=tip,x='day')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.displot(data=titanic,x='age', kind='hist',element='step',col='sex')
```

kdeplot

```python
sns.kdeplot(data=tip,x='total_bill')
```

```python
sns.displot(data=tip,x='total_bill', kind='kde',hue='sex')
```

rugplot

```python
sns.kdeplot(data=tip,x='total_bill')
sns.rugplot(data=tip,x='total_bill')
```

```python
#bivariate histplot
sns.histplot(data=tip,x='total_bill',y='tip')
```

```python
#bivariate kdeplot
sns.displot(data=tip,kind='kde',x='total_bill',y='tip')
```

# 3. Matrix Plot

*   Heatmap
*   Clustermap



heatmap

```python
gap.head()
```

```python
temp_df = gap.pivot_table(index='country',columns='year',values='lifeExp')
```

```python
# Here only axes level function
plt.figure(figsize=(10,10))
sns.heatmap(data=temp_df)
```

```python
temp_df = gap[gap['continent']=='Europe'].pivot_table(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(10,10))
# use annot for data also
sns.heatmap(data=temp_df,annot=True,cmap='viridis')
```

clustermap

```python
iris = sns.load_dataset('iris')
```

```python
sns.clustermap(data=iris.iloc[:,[0,1,2,3]],cmap='viridis')
```
